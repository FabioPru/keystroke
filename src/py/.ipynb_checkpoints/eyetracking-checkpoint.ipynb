```python
df = pd.read_csv('../../MonolingualReadingData.csv', sep=';')
s = pd.read_csv('../../sentences.csv', encoding='latin1')
```

```python
w = df[['PP_NR', 'WORD_ID', 'WORD', 'WORD_SKIP',
        'WORD_TOTAL_READING_TIME', 'WORD_SELECTIVE_GO_PAST_TIME', 'WORD_GAZE_DURATION', 
        'WORD_FIXATION_COUNT', 'WORD_RUN_COUNT',
        'WORD_FIRST_FIX_PROGRESSIVE', 'WORD_AVERAGE_FIX_PUPIL_SIZE', 'WORD_FIRST_FIXATION_DURATION'
        ]]
# TOTAL_READING_TIME = Summation of all fixation durations on the current word.
# SELECTIVE_GO_PAST = Summation of all fixation durations on the current word from when the current word is first fixated until the eyes enter a word with a higher word identification number.
# GAZE = Summation of all fixation durations in the first run on the current word.
```

```python
text = w[w['PP_NR'] == 'pp21']['WORD'].str.cat(sep=' ')

text_file = open("../../text.txt", "w")
text_file.write(text)
text_file.close()
```

```python
w = w[w['PP_NR'] != 'pp27']
w = w[w['PP_NR'] != 'pp25']
participants = ['pp32', 'pp22', 'pp26', 'pp29', 'pp28', 'pp23', 'pp31', 'pp34',
                'pp30', 'pp33', 'pp21', 'pp35']
```

```python
w.replace(to_replace='.', value=0, inplace=True)
w['WORD_AVERAGE_FIX_PUPIL_SIZE'] = w['WORD_AVERAGE_FIX_PUPIL_SIZE'].str.replace(',', '.')
w = w.astype({'WORD_TOTAL_READING_TIME': 'float64', 'WORD_SELECTIVE_GO_PAST_TIME': 'float64',
              'WORD_GAZE_DURATION': 'float64', 'WORD_FIRST_FIX_PROGRESSIVE': 'int64',
              'WORD_AVERAGE_FIX_PUPIL_SIZE': 'float64', 'WORD_FIRST_FIXATION_DURATION': 'float64'}, copy=True)
```

```python
a = w.groupby('WORD_ID').sum().reset_index()
a['WORD'] = w['WORD'].iloc[:len(a)]
cols = a.columns.tolist()
w = a[cols[:1] + cols[-1:] + cols[1:-1]]
```

```python
for idx in [3, 4, 5, 9, 10]:
    w.iloc[:, idx] = w.iloc[:, idx] / len(participants)
```

```python
# time: 20s
cpos = 0
for idx, row in w.iterrows():
    w.at[idx, 'pos'] = cpos
    cpos += len(row['WORD']) + 1
```

```python
# 5 mins
for idx, row in s.iterrows():
    w.loc[(w['pos'] >= row['start_pos']) & (w['pos'] <= row['end_pos']), 's_id'] = idx
```

```python
w
```

```python
w['t/len'] = w['WORD_TOTAL_READING_TIME'] / w['WORD'].apply(len)
```

```python
pupil = w.groupby('s_id')['WORD_AVERAGE_FIX_PUPIL_SIZE'].mean()
times = w.groupby('s_id')['t/len'].mean()
s_len = s['end_pos'] - s['start_pos']
```

```python
import textstat
```

```python
window = 200
ln = len(s) - window - 1

times_s = np.zeros(ln)
pupil_s = np.zeros(ln)
len_s = np.zeros(ln)

for i in range(0, ln):
    times_s[i] = times[i:i+window].mean()
    pupil_s[i] = pupil[i:i+window].mean()
    len_s[i] = s_len[i:i+window].mean()

mn = times_s.mean()
std = times_s.std()
times_s = (times_s - mn) / std

mn = pupil_s.mean()
std = pupil_s.std()
pupil_s = (pupil_s - mn) / std
```
