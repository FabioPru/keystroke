# Helper Functions

## Boxplot Helper Function

boxplot_helper(x_data,y_colname,data_in,n_quantiles,xlabel,ylabel) plots x = x_data and y = data_in[y_colname] in n_quantiles and returns the n_quantile bins. 

```python
def boxplot_helper(x_data,y_colname, data_in, n_quantiles,xlabel,ylabel):
    d_group,bins = pd.qcut(x_data,n_quantiles,duplicates = 'drop',retbins=True)
    ax = sns.boxplot(x=d_group, y=y_colname, data=data_in,showfliers=False)
    ax.set_xticks([])
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    print(bins)
```

## Plot Helper Function

```python
def plotDistributions(data1, data2, label1='data1', label2='data2', x_label='variable'):
    
    m = min(data1.quantile(0.01), data2.quantile(0.01))
    M = max(data1.quantile(0.99), data2.quantile(0.99))
    bins = np.linspace(m, M, 100)

    fig, axs = plt.subplots(2, gridspec_kw={'height_ratios': [1, 4]})
    fig.set_size_inches(20, 8)

    #plt.hold = True
    #boxes=[singleClause, multiClause]
    #axs[0].boxplot(boxes,vert=0)
    #axs[0].set(xlim=(m, M))

    axs[0].set(xlim=(m, M))
    my_pal = {0: "g", 1: "r"}
    sns.boxplot(data=[data1, data2],  orient='h' , ax=axs[0], palette=my_pal)

    axs[1].hist(data1, bins, alpha=0.4, label=label1, color='g')
    axs[1].hist(data2, bins, alpha=0.3, label=label2, color='r')
    axs[1].axvline(x=data1.mean(), c='g', lw=2)
    axs[1].axvline(x=data2.mean(), c='r', lw=2)
    axs[1].set(xlim=(m, M))


    plt.legend(loc='upper right')
    plt.xlabel(x_label)
    plt.show()
```

# Pause & Burst

```python
def get_burstlist(file_id,p_threshold):
    burstlst = []
    stsid = f.loc[file_id,'s_id_s']
    steid = f.loc[file_id,'s_id_e']
    k_in_f = k[(k['s_id']>=stsid)&(k['s_id']<=steid)]
    tmean = k_in_f['t'].mean()
    tmed = k_in_f['t'].median()
    prev = k_in_f.index[0]
    for i in range(len(k_in_f)):
        ks = k_in_f.iloc[i]
        if ks['t'] > p_threshold:
            if k_in_f.index[i]-1 < prev:
                burstlst.append((prev,prev,0))
            else:
                burstlst.append((prev,k_in_f.index[i],ks['t']))
            prev = k_in_f.index[i]
    return burstlst,tmean,tmed
```

```python
def visualize_burst(f_id_in,p_thresh_in):
    blist,tmean,tmed = get_burstlist(f_id_in,p_thresh_in)
    tot_str = ''
    for burst in blist:
        tot_str += ''.join(str(st).replace('nan','←') for st in k[burst[0]:burst[1]]['char'].values)
        tot_str += ' {'+str(burst[2])+'}'
    return tot_str
```

# Word Frequency vs Pause Length

```python
def word_info(word_len):
    wordlist = []
    freqlist = []
    tlist = []
    tmedlist=[]
    numlist = []
    allowed = 'abcdefghijklmnopqrstuvwxyz'
    same_len_words = w[(w['text_len'] == word_len) & (w['spell_err'] == 0)]
    gp = same_len_words.groupby(by='text')
    for name,group in gp:
        if all(c in allowed for c in name):
            wordfreq = group['word_freq'].iloc[0]
            t_avg = group['t/len_filter_10000'].mean()
            t_med = group['t/len_filter_10000'].median()
            wordlist.append(name)
            freqlist.append(wordfreq)
            tlist.append(t_avg)
            tmedlist.append(t_med)
            numlist.append(len(group))
    freq_t = pd.DataFrame()
    freq_t['word'] = wordlist
    freq_t['word frequency'] = freqlist
    freq_t['t'] = tlist
    freq_t['t_med'] = tmedlist
    freq_t['occ'] = numlist
    freq_t['log_t'] = np.log(1+freq_t['t'])
    freq_t['log_t_med'] = np.log(1+freq_t['t_med'])
    return freq_t
```

## Get Burst List (words)

get_burstlist_words2(file_id,p_threshold) takes in file id and pause threshold (ms) returns tuple list where the tuple is (word id of start of burst, word id of end of burst, pause length after burst)

```python
def get_burstlist_words2(file_id,p_threshold):
    burstlst = []
    stsid = f.loc[file_id,'s_id_s']
    steid = f.loc[file_id,'s_id_e']
#     sp_in_f = spaces[(spaces['s_id']>=stsid)&(spaces['s_id']<=steid)]
    sp_in_f = spaces[spaces['f_id']==file_id]
    tmean = sp_in_f['new_t'].mean()
    tmed = sp_in_f['new_t'].median()
    prev = sp_in_f.iloc[0]['w_before']
    for i in range(len(sp_in_f)):
        ks = sp_in_f.iloc[i]
        if ks['new_t'] > p_threshold:
            burstlst.append((prev+1,ks['w_before'],ks['new_t']))
            prev = ks['w_before']
    return burstlst,tmean,tmed
```
