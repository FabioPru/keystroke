```python
def splitByUser(cond1, cond2, tol=1, df=s):
    '''Execution time: approx. 13s
    
    Splits a dataframe into two, so that they have approximately the same user writer and text_len distribution
    cond1, cond2: functions row -> True/False
    tol: how strict should we be
    df: defaults into splitting sentences, can be changed at your risk'''
    
    global k, w, s, f, u
    s1 = []; s2 = []; len1 = 0; len2 = 0
    df.loc[:, 'new_len'] = (df['text_len'] / tol).astype(int)

    
    for i in range(0, len(u)):
        u_df = df[df['u_id'] == i]
        
        len_range = range(u_df['new_len'].min(), u_df['new_len'].max() + 1)
        d1 = {k: [] for k in len_range}; d2 = {k: [] for k in len_range}
        
        for idx, row in u_df.iterrows():
            if cond1(row):
                d1[row['new_len']].append(row['id'])
            if cond2(row):
                d2[row['new_len']].append(row['id'])
        
        for l in len_range:
            todo = min(len(d1[l]), len(d2[l]))
            s1 += d1[l][0:todo]
            s2 += d2[l][0:todo]
    
    ids = df['id']
    tv1 = df['id'] < 0; tv2 = df['id'] < 0   # all False
    s1p = {i: 1 for i in s1}; s2p = {i: 1 for i in s2}

    for idx, val in ids.iteritems():
        if val in s1p:
            tv1[idx] = True
        if val in s2p:
            tv2[idx] = True
    
    
    print("Returning dataframes of len ", len(df[tv1]))
    return df[tv1], df[tv2]
```

```python
#returns sequence list of sentence: e.g. list of [start_pos, end_pos] for each word
def plotSentence(s_id):
    #break up sentences by word
    i = 0
    sq_list = []
    for i in range(len(w)):
        if (w.loc[i, 's_id'] == s_id):
            print(i)
            f_id = w.loc[i, 'f_id']
            print(f_id)
            break
    while (w.loc[i, 's_id'] == s_id):
        sq_list.append([w.loc[i, 'start_pos'], w.loc[i, 'end_pos']])
        i += 1
    return (sq_list, f_id)
```

```python
#returns sequence list of file: e.g. list of [start_pos, end_pos] for each sentence or word
def plotFile(f_id, data_type):
    i = 0
    sq_list = []
    #break up file by word
    if (data_type == 's'):
        for i in range(len(s)):
            if (s.loc[i, 'f_id'] == f_id):
                break
        while(s.loc[i, 'f_id'] == f_id):
            sq_list.append([s.loc[i, 'start_pos'], s.loc[i, 'end_pos']])
            i += 1
    elif (data_type == 'w'):
        for i in range(len(w)):
            if (w.loc[i, 'f_id'] == f_id):
                print(i)
                break
        while(w.loc[i, 'f_id'] == f_id):
            sq_list.append([w.loc[i, 'start_pos'], w.loc[i, 'end_pos']])
            i += 1
    return sq_list
```

```python
#OFFICIAL VERSION
#plots average time per character by word or sentence, depending on sequence_list parameter
def plotSequenceFluency(f_id, sequence_list, func=(lambda x: x)):
    avg_insertion_time_by_section_list = []
    avg_deletion_time_by_section_list = []
    x_axis = list(range(len(sequence_list)))
    text_list = []
    #for each chunk of text
    for section in sequence_list:
        #append avg time per char (less than 10s) in chunk to list, by insertion and deletion
        section_insertion_mean = (k[(k.f_id == f_id) & (k.pos >= section[0]) & (k.pos <= section[1]) & (k.op == '+')].t.apply(func).mean())
        section_total_mean = (k[(k.f_id == f_id) & (k.pos >= section[0]) & (k.pos <= section[1])].t.apply(func).mean())
        section_deletion_mean = section_total_mean - section_insertion_mean
        avg_insertion_time_by_section_list.append(section_insertion_mean)
        avg_deletion_time_by_section_list.append(section_deletion_mean)
        #print(section_insertion_mean, section_deletion_mean, section_insertion_mean + section_deletion_mean)
        section_text = f.loc[f_id, 'text'][section[0]: section[1] + 1]
        text_list.append(section_text)
    #return avg_time_by_section_list
    plt.figure(figsize=(20, 15))
    p1 = plt.bar(x_axis, avg_insertion_time_by_section_list)
    p2 = plt.bar(x_axis, avg_deletion_time_by_section_list, bottom=avg_insertion_time_by_section_list)
    plt.xticks(x_axis, text_list, rotation='vertical')
    plt.ylabel("Time")
    plt.legend((p1[0], p2[0]), ('Insertions', 'Deletions'))
    #plt.yscale('log')
    
    return avg_insertion_time_by_section_list
```

```python
#plots word_t/t and separator_t/t in side-by-side bar graph
def wordtVsSeparatort(f_id):
    plt.figure(figsize=(15, 12))
    width = 0.35
    p1 = plt.bar(s[s.f_id == f_id].id, s[s.f_id == f_id]['word_t/t'], width)
    p2 = plt.bar(s[s.f_id == f_id].id + width, s[s.f_id == f_id]['separator_t/t'], width)
    plt.xlabel('index')
    plt.ylabel('percentage')
    plt.legend((p1[0], p2[0]), ('word_t/t', 'separator_t/t'))
```

```python
#plots any two sentence dataframe traits in side-by-side bar graph
def sideBySide(trait1, trait2, f_id):
    plt.figure(figsize=(15, 12))
    width = 0.35
    p1 = plt.bar(s[s.f_id == f_id].id, s[s.f_id == f_id][trait1], width)
    p2 = plt.bar(s[s.f_id == f_id].id + width, s[s.f_id == f_id][trait2], width)
    plt.legend((p1[0], p2[0]), (trait1, trait2))
```
