```python
def writeSuffixK (suffix=''):
    with open('../../pads/suffixK.txt', 'w') as suffixK:
        suffixK.write(suffix)
        suffixK.close()

def writeSuffixS (suffix=''):
    with open('../../pads/suffixS.txt', 'w') as suffixS:
        suffixS.write(suffix)
        suffixS.close()
        
def readSuffixK ():
    suffix = ''
    f = open('../../pads/suffixK.txt', 'r')
    suffix = f.read()
    f.close()
    return suffix

def readSuffixS ():
    suffix = ''
    f = open('../../pads/suffixS.txt', 'r')
    suffix = f.read()
    f.close()
    return suffix
```

```python
def loadKeystrokes (suffix=''):
    if suffix == '':
        suffix = readSuffixK()
    global k
    k = pd.read_csv('../../pads/keystrokes' + suffix + '.csv', index_col=False)

def loadSentences (suffix=''):
    if suffix == '':
        suffix = readSuffixK()
    global s
    s = pd.read_csv('../../pads/sentences' + suffix + '.csv')
```

```python
def writeKeystrokes (suffix=''):
    global k
    k.to_csv('../../pads/keystrokes' + suffix + '.csv', index=False)
    writeSuffixK(suffix)
    
def writeSentences (suffix=''):
    global s
    s.to_csv('../../pads/sentences' + suffix + '.csv', index=False)
    writeSuffixS(suffix)
```

## Add Fields: Subject ID, File ID

### Make a Dictionary of File's start, end s_id

```python
startid = 0
fid = 0
prevpath = ''
filedt = {}
fileids = []
sid2fileid = {}
for i in range(len(s)):
    path = s.loc[i,'path']
    if  i != 0 and path != prevpath:
        currsid = s.loc[i,'id']
        filedt[fid] = (startid,currsid-1)
        fid += 1
        startid = currsid
    sid2fileid[s.loc[i,'id']] = fid
    fileids.append(fid)
    prevpath = path
```

```python
s['file_id'] = fileids
```

### Add Subject ID to field & Make Dictionary of file Id and subject Id

```python
subid2path = {}
path2subid = {}
subjid = 0
for i in range(len(s)):
    idfrompath = s.loc[i,'path'].split('/')[2]
    if idfrompath not in path2subid:
        path2subid[idfrompath] = subjid
        subid2path[subjid] = idfrompath
        subjid += 1
```

```python
subjidlst = []
added = []
subjfiles = {k: [] for k in range(len(subid2path))}
subjsids = {k: [] for k in range(len(subid2path))}
fileid2subid = {}
for i in range(len(s)):
    idfrompath = s.loc[i,'path'].split('/')[2]
    subid = path2subid[idfrompath]
    subjidlst.append(subid)
    fileid = s.loc[i,'file_id']
    fileid2subid[fileid] = subid
    subjsids[subid].append(s.loc[i,'id'])
    if fileid not in added:
        subjfiles[subid].append(fileid)
        added.append(fileid)
```

```python
s['subj_id'] = subjidlst
```

## Counts of Characters

```python
counts = k['char'].value_counts().to_dict()
ctdict = {}
for ky in counts.keys():
    if (ky >= 'A' and ky <= 'Z') or (ky >= 'a' and ky <= 'z'):
        ctdict[ky] = counts[ky]
counts
```

## Pause length Mean & Median

```python
p_med = k['t'].median()
p_mean = k['t'].mean()
p_blank_med = k[k['char']==' ']['t'].median()
p_blank_mean = k[k['char']==' ']['t'].mean()
p_char_med = k[k['char']!=' ']['t'].median()
p_char_mean = k[k['char']!=' ']['t'].mean()
p_inst_med = k[k['op']=='+']['t'].median()
p_inst_mean = k[k['op']=='+']['t'].mean()
p_del_med = k[k['op']=='-']['t'].median()
p_del_mean = k[k['op']=='-']['t'].mean()
```

```python
rows = ['Overall','Blank','Alphabets','Insertion','Deletion']
mdata = [[p_mean,p_med],[p_blank_mean,p_blank_med], [p_char_mean,p_char_med],[p_inst_mean,p_inst_med],[p_del_mean,p_del_med]]
pd.DataFrame(mdata, columns=["Mean (ms)", "Median (ms)"], index = rows)
```
