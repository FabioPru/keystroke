```python
def getCharsRev(start_sent_id,end_sent_id,version):
    # version 0: first draft, version 1: second draft, ...
    letterlst = []
    stps = s.loc[start_sent_id,'start_pos']
    endps = s.loc[end_sent_id,'end_pos']
    kinfile = k[(k['s_id'] >= start_sent_id) & (k['s_id'] <= end_sent_id)]
    for i in range(stps,endps+1):
        kst = kinfile[k['pos']==i]
        num_revs = len(kst) - 1
        if num_revs < version:
            letter = ' '
        else:
            letter = kst.iloc[version]['char']
            if str(letter) == 'nan':
                letter = '←'
        letterlst.append((letter,num_revs))
    return letterlst
```

```python
start_sent_id = f.loc[0,'s_id_s']
end_sent_id = f.loc[0,'s_id_e']
firstdraft = getCharsRev(start_sent_id,end_sent_id,0)
seconddraft = getCharsRev(start_sent_id,end_sent_id,1)
```

# Pause & Burst

```python
def get_burstlist(file_id,p_threshold):
    burstlst = []
    stsid = f.loc[file_id,'s_id_s']
    steid = f.loc[file_id,'s_id_e']
    k_in_f = k[(k['s_id']>=stsid)&(k['s_id']<=steid)]
    tmean = k_in_f['t'].mean()
    tmed = k_in_f['t'].median()
    prev = k_in_f.index[0]
    for i in range(len(k_in_f)):
        ks = k_in_f.iloc[i]
        if ks['t'] > p_threshold:
            if k_in_f.index[i]-1 < prev:
                burstlst.append((prev,prev,0))
            else:
                burstlst.append((prev,k_in_f.index[i],ks['t']))
            prev = k_in_f.index[i]
    return burstlst,tmean,tmed
```

```python
blist,tmean,tmed = get_burstlist(100,600)
blist,_,_ = get_burstlist(100,2*tmed)
for burst in blist:
    print(''.join(str(st).replace('nan','←') for st in k[burst[0]:burst[1]]['char'].values),end='')
    print(' {'+str(burst[2])+'}',end='')
```

# Categorize Keystrokes

```python
k.head(5)
```

```python
prevfid = k.loc[0,'f_id']
for i in range(len(k)):
    currfid = k.loc[i,'f_id']
    if currfid == prevfid:
        if i< len(k) -1 and str(k.loc[i+1,'char']).isalpha() and str(k.loc[i,'char']).isalpha():
            k.at[i,'keyst_type'] = 0
        elif i< len(k) -1 and str(k.loc[i,'char']) == '.' and str(k.loc[i+1,'char']) == ' ':
            k.at[i,'keyst_type'] = 3
        elif i> 0 and str(k.loc[i,'char']) == ' ' and str(k.loc[i-1,'char']) == '.':
            k.at[i,'keyst_type'] = 4
        elif i < len(k)-1 and str(k.loc[i+1,'char']) == ' ':
            k.at[i,'keyst_type'] = 1
        elif str(k.loc[i,'char']) == ' ':
            k.at[i,'keyst_type'] = 2
    prevfid = currfid
```

```python
k_sorted = k.sort_values(by='f_id')
rs = 0; re = 0; ranges = {}
for idx, val in k_sorted['f_id'].value_counts().sort_index().iteritems():
    re += val
    ranges[idx] = (rs, re)
    rs += val
```

```python
for f_id, row in f.iterrows():
    if ranges.get(f_id) != None:
        u, v = ranges.get(f_id)
        k_file = k_sorted.iloc[u:v, :]
    else:
        k_file = k_sorted.iloc[0:0, :]
    #######
    text = f.at[f_id, 'text']
    for idx, row in k_file.iterrows():
        ps = row['pos']
        if ps<len(text)-1 and ps>0:
            if text[ps].isalpha() or text[ps].isdigit():
                if text[ps+1].isalpha() or text[ps+1].isdigit():
                    # inside a word
                    k.at[idx,'keyst_type'] = 0
                else:
                    # last char of a word
                    k.at[idx,'keyst_type'] = 1
            else:
                if text[ps-1].isalpha() or text[ps-1].isdigit():
                    if text[ps+1].isalpha() or text[ps+1].isdigit():
                        # space after word before word
                        k.at[idx,'keyst_type'] = 2
                    else:
                        # space after word before space
                        k.at[idx,'keyst_type'] = 3
                else:
                    # space after period/comma/etc.
                    k.at[idx,'keyst_type'] = 4                
        
        else:
            # first or last char of file
            k.at[idx,'keyst_type'] = -1  
```

```python
k.to_csv('../../pads/keystrokes-types.csv', index=False)
```

```python
lst = []
for i,row in w.iterrows():
    if w.at[i,'s_id'] == w.at[i+1,'s_id']:
        w1 = w.at[i,'text']
        w2 = w.at[i+1,'text']
        bg = str(w1)+' '+str(w2)
        spellerr = 1 if w.at[i,'spell_err'] == 1 or w.at[i+1,'spell_err'] == 1 else 0
        bigram_freq = bigr_freq[bg] if bg in bigr_freq else 0
        word_freq = w.at[i,'word_freq']*w.at[i+1,'word_freq']
        l1 = w.at[i,'text_len']
        l2 = w.at[i+1,'text_len']
        text_len = l1+l2
        k_count = w.at[i,'k_count']+w.at[i+1,'k_count']
        t = w.at[i,'t']+w.at[i+1,'t']
        t_over_len = t/text_len
        t_over_len_filter_2000 = (w.at[i,'t/len_filter_2000']*l1+w.at[i+1,'t/len_filter_2000']*l2)/text_len
        t_over_len_filter_10000 = (w.at[i,'t/len_filter_10000']*l1+w.at[i+1,'t/len_filter_10000']*l2)/text_len
        max_p = max(w.at[i,'max_p'],w.at[i+1,'max_p'])
        n_revisions = w.at[i,'n_revisions']+w.at[i+1,'n_revisions']
        rowdict = {'s_id': w.at[i,'s_id'],'f_id':w.at[i,'f_id'],'u_id':w.at[i,'u_id'],'text':bg,'start_pos' : w.at[i,'start_pos'],
                   'end_pos' : w.at[i+1,'end_pos'],'path' : w.at[i,'path'],
                   'spell_err':spellerr,'bigram_freq':bigram_freq,'word_freq':word_freq,'text_len':text_len,
                   'k_count':k_count,'t':t,'t/len':t_over_len,'t/len_filter_2000':t_over_len_filter_2000,
                   't/len_filter_10000':t_over_len_filter_10000,'max_p':max_p,'n_revisions':n_revisions}
        lst.append(rowdict)
```

```python
bi_gram = pd.DataFrame(data=lst, columns = ['s_id','f_id','u_id','text','start_pos','end_pos','path','spell_err','bigram_freq','word_freq','text_len','k_count','t','t/len','t/len_filter_2000','t/len_filter_10000','max_p','n_revisions'])
```
